```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv(r"C:\Users\avins\OneDrive\Desktop\Coffee Nutrient Deficiency\test_data.csv")

plt.figure()

# Create small vertical noise (jitter)
np.random.seed(42)

for label in df["Label"].unique():
    subset = df[df["Label"] == label]
    
    # Add small random noise for Y-axis
    jitter = np.random.normal(0, 0.02, len(subset))
    
    plt.scatter(subset["K"], jitter, label=label)

plt.xlabel("Potassium (K)")
plt.yticks([])   # Hide Y axis
plt.title("Clustering based only on K (Test Data)")
plt.legend()
plt.show()
```
