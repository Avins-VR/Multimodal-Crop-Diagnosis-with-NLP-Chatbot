```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv(r"C:\Users\avins\OneDrive\Desktop\Coffee Nutrient Deficiency\test_data.csv")

plt.figure()

# Small vertical jitter so points don’t overlap
np.random.seed(42)

for label in df["Label"].unique():
    subset = df[df["Label"] == label]
    
    jitter = np.random.normal(0, 0.02, len(subset))
    
    plt.scatter(subset["humidity"], jitter, label=label)

plt.xlabel("Humidity")
plt.yticks([])  # Hide Y-axis
plt.title("Clustering based only on Humidity (Test Data)")
plt.legend()
plt.show()
```
